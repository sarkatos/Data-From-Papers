```python
import requests
import os
import re
```

```python
# Especifica os valores de x e y para seu loop
x = 1604269  # Valor inicial
y = 1604874  # Valor final (ajuste conforme necessário)

# Nome de usuário e senha para autenticação
username = 'ztffps'
senha = 'dontgocrazy!'

# Diretório onde os arquivos .txt serão salvos
output_directory = "dados_recebidos_por_email"

# Certifica-se de que o diretório de saída exista
os.makedirs(output_directory, exist_ok=True)

for numero in range(x, y + 1):
    # URL do link com número variável   sim, 'ztfweb.ipac.caltech.edu' tá entre 2 barras //
    url = f"https://ztfweb.ipac.caltech.edu//ztf/ops/forcedphot/lc/batchfp/br007901-008000/7960/37d82bea8406d3c5413413fb2d2c1e58/batchfp_req000{numero}_lc.txt"

    auth = (username, senha)

    # Envia uma solicitação GET para o URL
    response = requests.get(url, auth=auth)

    if response.status_code == 200:
        # Extrai o R.A. e Dec. do conteúdo do arquivo
        content = response.text
        ra_match = re.search(r'Requested input R.A. = (.+) degrees', content)
        dec_match = re.search(r'Requested input Dec. = (.+) degrees', content)
        
        if ra_match and dec_match:
            ra = ra_match.group(1)
            dec = dec_match.group(1)
            
            # Use R.A. e Dec. para nomear o arquivo
            filename = f"ztf_ra{ra}_dec{dec}.txt"
            
            with open(os.path.join(output_directory, filename), "w", encoding="utf-8") as file:
                file.write(response.text)
            print(f"Arquivo {filename} salvo com sucesso.")
        else:
            print(f"Falha ao extrair R.A. e Dec. do arquivo {url}.")
    else:
        print(f"Falha ao acessar {url}. Status Code: {response.status_code}")
```
